```python
import numpy as np,pandas as pd
from sklearn.model_selection import cross_val_score
```

```python
df = pd.read_csv(r"C:\Users\hp\Downloads\data_cancer.csv")
```

```python
df.drop(columns= ['id','Unnamed: 32'],inplace= True)
```

```python
df
```

```python
df.columns
```

```python
x= df.iloc[:,1:]
y = df.iloc[:,0]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.iloc[:,1:],df.iloc[:,0],
                                                test_size= 0.2,random_state = 2)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors= 3)
```

```python
knn.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred = knn.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
cv_scores= cross_val_score(knn,x,y,cv = 5)
print(cv_scores)
```

```python
print('cv scores mean:{}'.format(np.mean(cv_scores)))
```

```python
scores = []
for i in range(1,11):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    y_pred= knn.predict(X_test)
    scores.append(accuracy_score(y_test,y_pred))
```

```python
import matplotlib.pyplot as plt   #k= 3 best accuracy
plt.plot(range(1,11),scores)
```
